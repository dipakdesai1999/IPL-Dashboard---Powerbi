```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
## Relational plot-> figure level 


## Line and scatter
```

```python
tips =sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
sns.relplot(data=tips, x='total_bill',y='tip',kind='scatter', hue='sex', style='time',size='size')
```

```python
gap = px.data.gapminder()
```

```python
gap
```

```python
temp_df = gap[gap['country']=='India']
```

```python
temp_df
```

```python
sns.lineplot(data=temp_df, x='year',y='lifeExp')
```

```python
sns.relplot(data=temp_df, x='year',y='lifeExp',kind='line')
```

```python
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp',hue='country')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip',kind='line',col='sex',row='day')
```

```python
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', col='year',col_wrap=3)
```

```python
## DIstribution plot- >figure level

## histplot, kdeplot
```

```python
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='day')
```

```python
sns.displot(data=tips,x='tip',hue='sex')
```

```python
sns.displot(data=tips,x='tip',hue='sex',element='step')
```

```python
sns.displot(data=tips,x='tip',col='sex',element='step')
```

```python
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
temp_df= gap.pivot(index='country', columns='year', values='lifeExp')
temp_df
```

```python
sns.heatmap(temp_df)
```

```python
temp_df = gap[gap['continent']=='Asia'].pivot(index='country', columns='year', values='lifeExp')
temp_df
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5)
```
