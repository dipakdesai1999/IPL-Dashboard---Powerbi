```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv("historical_automobile_sales.csv")
sales
```

Q 1: Develop a Line chart using the functionality of pandas to show how
automobile sales fluctuate from year to year. 

```python
yearly_sales = sales.groupby("Year")["Automobile_Sales"].sum().reset_index()
plt.plot(yearly_sales["Year"], yearly_sales["Automobile_Sales"], marker="o", linestyle="--", color="b",label = "Automobile_Sales")
plt.xlabel("Year")
plt.ylabel("Total Automobile Sales")
plt.title("Yearly Automobile Sales Fluctuation")
plt.legend()
plt.grid(True)
```
