```python
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],label='Virat Kohli',marker='o',markersize=8)
plt.plot(batsman['index'],batsman['RG Sharma'],label='Rohit Sharma',marker='*',markersize=10)
plt.legend()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],label='Virat Kohli',marker='o',markersize=8)
plt.plot(batsman['index'],batsman['RG Sharma'],label='Rohit Sharma',marker='*',markersize=10)
plt.title("Rohit VS Virat Career Comparision")
plt.xlabel('Season')
plt.ylabel("Runs Scored")
plt.legend()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],label='Virat Kohli',marker='D',markersize=8,linestyle='dashed',linewidth=2,color='purple')
plt.plot(batsman['index'],batsman['RG Sharma'],label='Rohit Sharma',marker='o',markersize=10,linestyle='dashdot',linewidth=3)
plt.title("Rohit VS Virat Career Comparision")
plt.xlabel('Season')
plt.ylabel("Runs Scored")
plt.legend()
plt.show()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],label='Virat Kohli',marker='D',markersize=8,linestyle='dashed',linewidth=2,color='purple')
plt.plot(batsman['index'],batsman['RG Sharma'],label='Rohit Sharma',marker='o',markersize=10,linestyle='dashdot',linewidth=3)
plt.title("Rohit VS Virat Career Comparision")
plt.xlabel('Season')
plt.ylabel("Runs Scored")
plt.legend(loc='upper right')

for i in range (batsman.shape[0]):
    plt.text(batsman['index'].values[i],batsman['V Kohli'].values[i],batsman['V Kohli'].values[i])
    plt.text(batsman['index'].values[i],batsman['RG Sharma'].values[i],batsman['RG Sharma'].values[i])
plt.grid()
plt.tex
```

right lower' is not a valid value for loc; supported values are 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'

```python
df = pd.read_csv('batsman_season_record.csv')
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='lightblue',label='2015')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='purple',label='2016')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='gray',label='2017')
plt.legend()
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
plt.bar(df['batsman'],df['2017'],label='2017',color='gray')
plt.bar(df['batsman'],df['2016'],label='2016',bottom=df['2017'],color='purple')
plt.bar(df['batsman'],df['2015'],label='2015',bottom=df['2017']+df['2016'],color='lightblue')
```
